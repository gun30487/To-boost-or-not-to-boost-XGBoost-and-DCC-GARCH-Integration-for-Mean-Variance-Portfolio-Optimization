```python
#  CHECK COLUMN NAME
stocks = ['EZU','EWA','EWJ','EWU','IVV','MCHI','INDA','VNM','THD']
col_diff = ['gold', 'BDI', 'BRT','EUR/USD','AUD/USD','JPY/USD','GBP/USD','USD','CNY/USD','INR/USD','VND/USD','THB/USD']
df = pd.DataFrame(index=range(1, 21))
for stock in stocks:
    macro = pd.read_excel(stock+'_RAW_Macro_D.xlsx')
    macro.rename(columns={'Unnamed: 0':'Date'},inplace=True)
    macro = macro.set_index('Date')
    for col in macro.columns :
        if col in col_diff :
            macro[col+'_D_diff'] = ((macro[col]/macro[col].shift(1))-1)*100
            macro[col+'_M_diff'] = ((macro[col]/macro[col].shift(30))-1)*100
            macro.drop(col, axis=1, inplace=True)
    macro.dropna(inplace=True)
    file_name = stock + '_Macro_D'
    macro.to_excel(file_name + '.xlsx')
    print(file_name+' DONE')
```
