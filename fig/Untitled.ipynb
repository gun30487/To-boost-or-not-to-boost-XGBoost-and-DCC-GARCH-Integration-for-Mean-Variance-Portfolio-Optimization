```python
from PIL import Image
import matplotlib.pyplot as plt

periods = ['2021-01','2022-01','2023-01','2024-01','2025-01']
for period in periods:
    img1 = Image.open(period+'_EFF_Benchmark.png')
    img2 = Image.open(period+'_EFF_HR+DCC.png')
    img3 = Image.open(period+'_EFF_XGB+DCC.png')

    images = [img1, img2, img3]
    
    # Create subplot grid
    fig, axes = plt.subplots(5, 3, figsize=(20, 6))
    
    for ax, img in zip(axes.flatten(), images):
        ax.imshow(img)
        ax.axis('off')
    
plt.tight_layout()
plt.savefig(period+'_EFF.png', dpi=300)
plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt

periods = ['2021-01','2022-01','2023-01','2024-01','2025-01']

fig, axes = plt.subplots(5, 3, figsize=(18, 20))

for i, period in enumerate(periods):

    img1 = Image.open(period + '_EFF_Benchmark.png')
    img2 = Image.open(period + '_EFF_HR+DCC.png')
    img3 = Image.open(period + '_EFF_XGB+DCC.png')

    images = [img1, img2, img3]

    for j, img in enumerate(images):

        ax = axes[i, j]

        ax.imshow(img)
        ax.axis('off')

plt.tight_layout()

plt.savefig('ALL_EFF.png', dpi=300, bbox_inches='tight')

plt.show()
```
